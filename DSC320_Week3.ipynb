```python
# c
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression


# import the data
nd = pd.read_csv('us_pop_data.csv')

# create a new column that is number of years since 1790
nd['years_since_1790'] = nd['year'] - 1790

# create a new column that is the population in millions
nd['pop_millions'] = nd['us_pop'] / 1000000

# plot the data
plt.plot(nd['year'], nd['pop_millions'], marker='o', linestyle='-')

# set the labels
plt.xlabel('Year')
plt.ylabel('Population (in millions)')

# split the data
x = nd['year'].to_numpy().reshape(-1, 1) 
y = nd['pop_millions']

# create a linear regression model
model = LinearRegression()

# train the model on the data
model.fit(x, y)

# estimating the population growth
y_pred = model.predict(x)

# R^2 value
r2 = y_pred ** 2

print(f'Linear regression model predictions: {y_pred}')
print()
print(f'R-squared value: {r2}')

# plot regression
plt.plot(x, y_pred, color='red')

# show plot
plt.show()

```

```python
# e
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression


# import the data
nd = pd.read_csv('us_pop_data.csv')

# create a new column that is number of years since 1790
nd['years_since_1790'] = nd['year'] - 1790

# create a new column that is the population in millions
nd['pop_millions'] = nd['us_pop'] / 1000000

# create a new column that is the number of years since 1790 squared
nd['years_since_squared'] = nd['years_since_1790'] ** 2 

# plot the data
plt.plot(nd['years_since_squared'], nd['pop_millions'], marker='o', linestyle='-')

# set the labels
plt.xlabel('Year')
plt.ylabel('Population (in millions)')

# split the data
x = nd['years_since_squared'].to_numpy().reshape(-1, 1) 
y = nd['pop_millions']

# create a linear regression model
model = LinearRegression()

# train the model on the data
model.fit(x, y)

# estimating the population growth
y_pred = model.predict(x)

# R^2 value
r2 = y_pred ** 2

print(f'Linear regression model predictions: {y_pred}')
print()
print(f'R-squared value: {r2}')

# plot regression
plt.plot(x, y_pred, color='red')

# show plot
plt.show()
```

```python
# f
# The linear regression for the squared of the number of years since 1790 fits the data better than the 
# linear regression for years from 1790.
# Yes it is apparent in the R^2-values because you can see how close those values are to the population values in the data.
```

```python
# b and c
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# importing the dataset
data = pd.read_csv('customer_spending.csv')

# apply log transformation 
log_data = np.log(data)

# plot histogram
plt.hist(log_data['ann_spending'], edgecolor='black', label='Log Transformed Data')

# add labels
plt.xlabel('Annual Spending')
plt.ylabel('Frequency')
plt.legend()

# show the plot
plt.show()
```
